```python
import numpy as np
import time
print("¡Hola!")
time.sleep(1)
nombre = (input("Bienvenido, por favor ingresa tu nombre: "))
print(f"Bienvenido, {nombre}, debes hundir los 3 barcos ocultos, para ello debes ingresar la fila y luego la columna en la cual vas a disparar")
# tablero
tamano_tablero = 5
tablero = np.full((tamano_tablero, tamano_tablero), "O", dtype=str)
# Funciónes
def imprimir_tablero(tablero):
    print(tablero)
def posicion_aleatoria(tamano_tablero):
    return np.random.randint(tamano_tablero)
# Longitudes
longitud_barco1 = 1
longitud_barco2 = 2
longitud_barco3 = 3
# Posicion
fila_barco1, col_barco1 = posicion_aleatoria(tamano_tablero), posicion_aleatoria(tamano_tablero)
# no superponer
while True:
    fila_barco2, col_barco2 = posicion_aleatoria(tamano_tablero), posicion_aleatoria(tamano_tablero)
    if fila_barco2 != fila_barco1 or col_barco2 != col_barco1:
        break
while True:
    fila_barco3, col_barco3 = posicion_aleatoria(tamano_tablero), posicion_aleatoria(tamano_tablero)
    if (fila_barco3 != fila_barco1 or col_barco3 != col_barco1) and (fila_barco3 != fila_barco2 or col_barco3 != col_barco2):
        break
# Inicio
while np.any(tablero == "O"):
    imprimir_tablero(tablero)
    try:
        time.sleep(1)
        fila_adivinanza = int(input("Adivina la fila: "))
        col_adivinanza = int(input("Adivina la columna: "))
    except ValueError:
        print("Error: Ingresa solo números enteros.")
        continue
# Disparos
    if ((fila_adivinanza == fila_barco1 and col_adivinanza == col_barco1) or
        (fila_adivinanza == fila_barco1 and col_adivinanza == col_barco1 + 1) or
        (fila_adivinanza == fila_barco1 and col_adivinanza == col_barco1 + 2) or
        (fila_adivinanza == fila_barco2 and col_adivinanza == col_barco2) or
        (fila_adivinanza == fila_barco2 and col_adivinanza == col_barco2 + 1) or
        (fila_adivinanza == fila_barco3 and col_adivinanza == col_barco3) or
        (fila_adivinanza == fila_barco3 and col_adivinanza == col_barco3 + 1) or
        (fila_adivinanza == fila_barco3 and col_adivinanza == col_barco3 + 2)):
        print("Tu disparo ha acertado")
# colocar X
        tablero[fila_adivinanza, col_adivinanza] = "X"
    else:
# Verificar disparo en el tablero
        if (fila_adivinanza < 0 or fila_adivinanza >= tamano_tablero
            or col_adivinanza < 0 or col_adivinanza >= tamano_tablero):
            print("coordenada fuera del tablero.")
        else:
            print("Disparo")
            tablero[fila_adivinanza, col_adivinanza] = "."
print("Has ganado")
```
